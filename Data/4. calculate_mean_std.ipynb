```python
import matplotlib.pyplot as plt
import cv2
from tqdm import tqdm

# check if when we read the images using cv2 we have all channels equal
for img in tqdm(os.listdir("./Original Images/Train")):
    # Load the image
    image = cv2.imread(os.path.join("./Original Images/Train", img))
    if (image[:,:,0] == image[:,:,1]).all() and (image[:,:,1] == image[:,:,2]).all():
      continue
    else:
      print("error")
```

```python
# calculate the standard deviation
import cv2
import numpy as np
import os

def global_std_dev(image_paths):
    n_total = 0
    mean_sum = 0
    m2 = 0  # Sum of squares of differences from the current mean

    for image_path in tqdm(image_paths):
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        n = img.size
        mean = np.mean(img)
        variance = np.var(img)

        n_total += n
        mean_sum += mean * n
        m2 += variance * (n - 1) + n * (mean ** 2)

    global_mean = mean_sum / n_total
    global_variance = (m2 - n_total * global_mean ** 2) / (n_total - 1)
    global_std_dev = np.sqrt(global_variance)
    print(global_mean)

    return global_std_dev

# Example usage
image_dir = "./Original Images/Train"
image_paths = [os.path.join(image_dir, img) for img in os.listdir(image_dir) if img.endswith('.png')]  # Adjust the extension as needed
std_dev = global_std_dev(image_paths)
print(f'Global Standard Deviation of the dataset: {std_dev}')
```
